## 2) Text Preprocessing

#### 1) Removing Punctuations
#### 2) Tokenization
#### 3) Stemming
#### 4) Removing Stopwords

```python
import string
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stopwords=stopwords.words('english')
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in stopwords]
    return text
```

### Converting text into vectors using Tf-idf Vctorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(analyzer=clean_text)
X_tfidf = tfidf_vect.fit_transform(df['Consumer complaint narrative'])
```

## 3) Selection Of Multiclass Classification Model

### 3.1) Splitting Dataset to compare different models
#### 6 Train dataset and 6 test dataset for our 6 models

```python
fraction=0.16
point=int(fraction*X_tfidf.shape[0])
print(point)
```

```python
y=df['Category']
```

```python
X_dt=X_tfidf[:point,:]
y_dt=y.iloc[:point]
X_knn=X_tfidf[point:2*point,:]
y_knn=y.iloc[point:2*point]
X_nb=X_tfidf[2*point:3*point,:]
y_nb=y.iloc[2*point:3*point]
X_sgd=X_tfidf[3*point:4*point,:]
y_sgd=y.iloc[3*point:4*point]
X_rf=X_tfidf[4*point:5*point,:]
y_rf=y.iloc[4*point:5*point]
X_em=X_tfidf[5*point:,:]
y_em=y.iloc[5*point:]
```

### 3.2) Decision Tree Classifier Model

```python
# Fit a basic model, make predictions on the holdout test set, and the generate the evaluation metrics
from sklearn.tree import DecisionTreeClassifier
X_train_dt,X_test_dt,y_train_dt,y_test_dt=train_test_split(X_dt,y_dt,test_size=0.2,random_state=1)

dt = DecisionTreeClassifier()
dt.fit(X_train_dt, y_train_dt)

y_pred_dt = dt.predict(X_test_dt)

print(classification_report(y_test_dt, y_pred_dt))
```

### 3.3) K Nearest Neighbors Classifier Model

```python
# Fit a basic model, make predictions on the holdout test set, and the generate the evaluation metrics
from sklearn.neighbors import KNeighborsClassifier
X_train_knn,X_test_knn,y_train_knn,y_test_knn=train_test_split(X_knn,y_knn,test_size=0.2,random_state=1)

knn=KNeighborsClassifier()
knn.fit(X_train_knn, y_train_knn)

y_pred_knn = knn.predict(X_test_knn)

print(classification_report(y_test_knn, y_pred_knn))
```

### 3.4) Naive Bayes Classifier

```python
# Fit a basic model, make predictions on the holdout test set, and the generate the evaluation metrics
from sklearn.naive_bayes import MultinomialNB
X_train_nb,X_test_nb,y_train_nb,y_test_nb=train_test_split(X_nb,y_nb,test_size=0.2,random_state=1)
nb = MultinomialNB()
nb.fit(X_train_nb, y_train_nb)

y_pred_nb = nb.predict(X_test_nb)

print(classification_report(y_test_nb, y_pred_nb))
```

### 3.5) Stochastic Gradient Descent (SGD) Classifier

```python
# Fit a basic model, make predictions on the holdout test set, and the generate the evaluation metrics
from sklearn.linear_model import SGDClassifier
X_train_sgd,X_test_sgd,y_train_sgd,y_test_sgd=train_test_split(X_sgd,y_sgd,test_size=0.2,random_state=1)
sgd = SGDClassifier()
sgd.fit(X_train_sgd, y_train_sgd)

y_pred_sgd = sgd.predict(X_test_sgd)

print(classification_report(y_test_sgd, y_pred_sgd))
```

### 3.6) Random Forest Classifier

```python
# Fit a basic model, make predictions on the holdout test set, and the generate the evaluation metrics
from sklearn.ensemble import RandomForestClassifier
X_train_rf,X_test_rf,y_train_rf,y_test_rf=train_test_split(X_rf,y_rf,test_size=0.2,random_state=1)
rf = RandomForestClassifier()
rf.fit(X_train_rf, y_train_rf)

y_pred_rf = rf.predict(X_test_rf)

print(classification_report(y_test_rf, y_pred_rf))
```

### 3.7) Ensemble Voting Classifier Model
#### Ensemble model by combining Decision Tree, KNN, NB, SGD, RF

```python
# Fit a basic model, make predictions on the holdout test set, and the generate the evaluation metrics
from sklearn.ensemble import VotingClassifier
X_train_em,X_test_em,y_train_em,y_test_em=train_test_split(X_em,y_em,test_size=0.2,random_state=1)
em = VotingClassifier(estimators=[('dt',dt),('knn',knn),('nb',nb),('sgd',sgd),('rf',rf)], voting='hard', weights=[1,1,1,3,2])
em.fit(X_train_em, y_train_em)

y_pred_em = em.predict(X_test_em)

print(classification_report(y_test_em, y_pred_em))
```

## 4) Comparison Of Models

```python
accuracy={'DT':accuracy_score(y_test_dt, y_pred_dt),
          'KNN':accuracy_score(y_test_knn, y_pred_knn),
          'NB':accuracy_score(y_test_nb, y_pred_nb),
          'SGD':accuracy_score(y_test_sgd, y_pred_sgd),
          'RF':accuracy_score(y_test_rf, y_pred_rf),
          'EM':accuracy_score(y_test_em, y_pred_em)}
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract keys (models) and values (accuracies)
models = list(accuracy.keys())
accuracies = list(accuracy.values())

# Create a colormap (coolwarm) and adjust the normalization scale
colors = plt.cm.coolwarm((np.array(accuracies) - min(accuracies)) / (max(accuracies) - min(accuracies)))

# Create a bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(models, accuracies, color=colors)
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy of Different Models')
plt.ylim(0, 1)  # Set the y-axis limit between 0 and 1 for accuracy values
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

# Add color bar for reference
sm = plt.cm.ScalarMappable(cmap=plt.cm.coolwarm, norm=plt.Normalize(vmin=min(accuracies), vmax=max(accuracies)))
sm.set_array([])  # Empty array
cbar = plt.colorbar(sm, orientation='vertical')
cbar.set_label('Accuracy Scale')

# Add annotations with accuracy values
for bar, acc in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, acc + 0.01, f'{acc:.2f}', ha='center')

plt.show()
```
